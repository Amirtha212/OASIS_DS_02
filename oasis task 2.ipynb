```python
df=pd.DataFrame(df)
x=df['Region']
y=df[' Estimated Unemployment Rate (%)']
plt.figure(figsize=(10, 6))
plt.title('Unemployment Rate Over Time')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
